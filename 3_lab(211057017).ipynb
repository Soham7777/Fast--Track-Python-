# DATA TYPE - STRING

```python

string = "Hello world"

print(string)

print ("\nThe first character is:",string[0]) #printing the first character

print("\nCharacters from 1 to 3 are",string[1:3]) #printing character in range(upper bound exclusive)

print("\nPrinting string 2 times:",string * 2) #printing string 2 times

print("\nUppercase is:",string.upper()) #all in uppercase

print("\nLowercase is:",string.lower()) #all in lowercase

print("\nAfter changing lowercase to uppercase and vice-versa:",string.swapcase()) #vice versa of lower and uppercase

print("\nIs there numerice value or not?:",string.isdigit()) #checks for numeric values

print("\nConcatenation :",string + "hello") #concatenation

print("\nAll the characters are in uppercase?",string.isupper()) #checikng if all characters are uppercase

print("\nAll the characters are in lowercase?",string.islower()) #checking if all characters are lowercase

```

# DATATYPE - TUPLES

```python
tuple = ("hello",1,2,1.2,"abc")

tuple_2 = ("xyz",3,4,50.40,"pqr")



print("\n Original Tuple:" ,tuple)

print("\n First Element:",tuple[0])  #Printing the First Element

print("\n Concatenation:",tuple + tuple_2) #concatenation

print("\n 2 times tuple:", tuple * 2)


# Nested tuple

tuple_3 = (tuple,tuple_2)

print("\n",tuple_3)

#Slicing operation on tuple


print("\n",tuple[::-1]) #reversing the tuple

print("\n",tuple[1:]) #printing from 2 index

print("\n",tuple[:-1]) #not printing the last element


#tuple deletion

del tuple_2 # cannot modify the tuple only can delete the whole tuple


#list in the tuple can be modified 

tuple_4 = (1,2,3,4,[5,6])

tuple_4[4][0] = 7

print("\n Modified tuple: ",tuple_4)

print(type(tuple_4))
```

# DATATYPE - SET[]


```python
set = {"hello","orange","xyz"}

set_2 = {"world","orange","apple"}

print("\n Original set: ",set)

#adding elements to set

set.add("abc")

print("\n added to set:",set)



#returning copy of a set

print("\n Copy of set",set.copy())


#removing specified item

set_2.discard("apple")

print("\n Removing one item:",set_2)



#Set containing similarities between two or more sets

intersection_set = set.intersection(set_2)

print("\n intersection :",intersection_set)



#Removing item exsting in both list

set.difference_update(set_2)

print("\n Removing the item existing in both :",set)



#symmetric difference returns a set that contains all items from both set but no common items

symmetric = set.symmetric_difference(set_2)

print("\n After symmetric operation:", symmetric)


#updating set with another set or element
y = {1,2,3}

set.update(y)

print("\n Updated set :",set)



#union of set

x = {1,2,3,5}
y = {1,3,6,7,}

print("\n original sets:",x,y)
union = x.union(y)

print("\n Union of sets is :",union)

#Checking for superset

z = x.issuperset(y)

print("\n Superset check:",z)

#frozen set creates an immutable set

fruits = {'apple','bananas','orange'}

f_set = frozenset(fruits)

print("\n",f_set)

```
